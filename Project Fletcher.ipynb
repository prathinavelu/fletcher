```python
removeUseless = compile(r'([\s]@[\S]*)|([\s]&amp[\S]*)|([\s]http[\S]*)|(^#[\S]*])|(^@[\S]*)|(\b[\S]{1}\b)')
specialChar = compile('[^a-zA-Z0-9 ]+')

def filterTweet(tweet):
    filteredTweet = specialChar.sub('',removeUseless.sub('', tweet)).lower().replace(' rt ',' ').replace('icymi','').replace(' mt ',' ')
    return filteredTweet
```

```python
tweet_cursor = congress.find({},{'user','text'})
congress_dict = defaultdict(list)
for t in tweet_cursor:
    if t['user'] in congressmen:
        text = filterTweet(t['text'])
        if text[:2] == 'rt':
            text = text[2:]
        congress_dict[t['user']].append(text)
```

```python
for k in congress_dict.keys():
    congress_dict[k] = ' '.join(congress_dict[k])
```

```python
vectorizer = TfidfVectorizer(stop_words="english", ngram_range=(1,2),max_features=2000,max_df=.7,min_df=.4)
tweet_vectors = vectorizer.fit_transform(congress_dict.values())
#tweet_vectors = vectorizer.fit_transform(tweets)
tweet_features = vectorizer.get_feature_names()
```

```python
clust_num = 4

k_means = cluster.KMeans(n_clusters=clust_num,init='k-means++',n_init=1000)
k_means.fit(tweet_vectors) 
congress_clusters = k_means.predict(tweet_vectors)

# db = cluster.DBSCAN(eps=0.8, min_samples=2)
# congress_clusters = db.fit_predict(tweet_vectors.toarray())

# agg = cluster.AgglomerativeClustering(n_clusters=clust_num, affinity='cosine',linkage='complete')
# congress_clusters = agg.fit_predict(tweet_vectors.toarray())
```
